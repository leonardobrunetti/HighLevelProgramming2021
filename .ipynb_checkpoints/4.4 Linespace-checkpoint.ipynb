```python
import math as m
import numpy as np
```

## Creating the array

```python
a = np.linspace(0, (2*np.pi), 100)
a
```

## Extracting every 10th element

```python
b = a[::10]
b
```

## Reversing the array

```python
c = a[::-1]
c
```

## Extracting elements for which difference between sin and cos is < 0.1

```python
mask = (abs(np.sin(a) - np.cos(a) < 0.1))
a[mask]       
```
